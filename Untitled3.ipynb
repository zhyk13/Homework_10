```python
import pandas as pd
import random

# Функция получает строку и возвращает True
# если эта строка равна 'robot'. В противном
# случае, функция возвращает False.
def category(who: str) -> bool:
  if who == 'robot':
    return True
  else:
    return False

# В список lst записываем 10 элементов 'robot'
lst = ['robot'] * 10
# В список lst добавляем 10 элементов 'human'
lst += ['human'] * 10
# В списке lst cлучайным образом 'перемешиваем'
# элементы 'robot' и 'human'
random.shuffle(lst)
# Из списка lst создаем DataFrame data
data = pd.DataFrame({'whoAmi':lst})
# Создаем стобец 'robot'  и заполняем его значениями
# True если в соответствующей строке столбца 'whoAmi'
# содержится значение 'robot', и значениями 'False,
# если в соответствующей строке столбца 'whoAmi'
# содержится значение 'human'
data['robot'] = [category(i) for i in data['whoAmi']]
# Создаем стобец 'human'  и заполняем его значениями
# True если в соответствующей строке столбца 'whoAmi'
# содержится значение 'human', и значениями 'False,
# если в соответствующей строке столбца 'whoAmi'
# содержится значение 'robot'
data['human'] = [not(category(i)) for i in data['whoAmi']]
# Считываем значение получившегося дата фрейма для проверки
data.head(20)
```
